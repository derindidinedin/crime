## Categorising the Data

```python
# Defining non-promotion periods
non_promo_periods = [
    (pd.to_datetime('2015-01-01'), pd.to_datetime('2015-02-09')),
    (pd.to_datetime('2015-02-18'), pd.to_datetime('2015-03-14')),
    (pd.to_datetime('2015-03-23'), pd.to_datetime('2015-05-23')),
    (pd.to_datetime('2015-06-02'), pd.to_datetime('2015-06-20')),
    (pd.to_datetime('2015-06-29'), pd.to_datetime('2015-08-01'))
]

# Filtering non-promotion periods
non_promo_sales = pd.concat(
    [sales_data[(sales_data['Date'] >= start) & (sales_data['Date'] <= end)] for start, end in non_promo_periods]
)

# Calculate weekly sales
non_promo_sales['Week'] = non_promo_sales['Date'].dt.isocalendar().week
weekly_sales = non_promo_sales.groupby(['ProductCode', 'StoreCode', 'Week'])['SalesQuantity'].sum().reset_index()

# Calculate average weekly sales per store for each product
avg_weekly_sales = weekly_sales.groupby(['ProductCode', 'StoreCode'])['SalesQuantity'].mean().reset_index()

# Clustering products
product_avg_sales = avg_weekly_sales.groupby('ProductCode')['SalesQuantity'].mean()
product_labels = pd.qcut(product_avg_sales, 3, labels=['Slow', 'Medium', 'Fast'])
product_clusters = product_avg_sales.reset_index().assign(Cluster=product_labels)

# Clustering stores
store_avg_sales = avg_weekly_sales.groupby('StoreCode')['SalesQuantity'].mean()
store_labels = pd.qcut(store_avg_sales, 3, labels=['Slow', 'Medium', 'Fast'])
store_clusters = store_avg_sales.reset_index().assign(Cluster=store_labels)

# Display the clusters
display(product_clusters.head())
display(store_clusters.head())
```

```python
import pandas as pd
import numpy as np

# Load all datasets
file_paths = {
    'sales_data': '/assignment4.1a.csv',
    'additional_sales_data': '/assignment4.1b.csv',
    'product_categories': '/assignment4.1c.csv',
    'promotion_dates': '/PromotionDates.csv'
}

dataframes = {name: pd.read_csv(path) for name, path in file_paths.items()}

# Replace '-' with NaN for all columns and convert to appropriate data types
for name, df in dataframes.items():
    df.replace('-', np.nan, inplace=True)

# Function to convert a column to datetime and identify missing dates
def convert_to_datetime(df, column):
    df[column] = pd.to_datetime(df[column], errors='coerce')
    missing_dates = df[column].isnull().sum()
    if missing_dates > 0:
        print(f"There are {missing_dates} missing dates in column {column}.")
    return df

# Convert date columns to datetime
dataframes['sales_data'] = convert_to_datetime(dataframes['sales_data'], 'Date')
dataframes['additional_sales_data'] = convert_to_datetime(dataframes['additional_sales_data'], 'Date')
dataframes['promotion_dates'] = convert_to_datetime(dataframes['promotion_dates'], 'StartDate')
dataframes['promotion_dates'] = convert_to_datetime(dataframes['promotion_dates'], 'EndDate')

# Remove duplicates
for name, df in dataframes.items():
    df.drop_duplicates(inplace=True)

# Display the cleaned data
for name, df in dataframes.items():
    print(f"\n{name} after cleaning:")
    display(df.head())
```
